````python
# Jupyter Notebook ÏÖÄÏóêÏÑú FASTA ÏûÖÎ†• Ï†ÄÏû•
import pandas as pd

df = pd.read_csv("../data/processed/positive_windows.csv")

with open("../data/processed/tmp_rnafold_input.fa", 'w') as f:
    for i, seq in enumerate(df["sequence"]):
        f.write(f">seq{i}\n{seq}\n")
````

````python
# ÌîÑÎ°úÏ†ùÌä∏ Î£®Ìä∏ ÎîîÎ†âÌÜ†Î¶¨ Í∏∞Ï§Ä
#!cd ~/SNU/2025-1/bioinfo1/LIN28A_project

# RNAfold Ïã§Ìñâ (dot-bracket Íµ¨Ï°∞ ÏòàÏ∏° Í≤∞Í≥ºÎ•º output.txtÏóê Ï†ÄÏû•)
#!RNAfold < data/processed/tmp_rnafold_input.fa > data/processed/structure_output.txt
````

````python
# structure_output.txtÏóêÏÑú dot-bracket Íµ¨Ï°∞ ÌååÏã±
dots = []
with open("../data/processed/structure_output.txt") as f:
    lines = f.readlines()
    for i in range(2, len(lines), 3):
        dot = lines[i].split()[0]
        dots.append(dot)

# Í∏∏Ïù¥ ÏùºÏπò Ïó¨Î∂Ä ÌôïÏù∏
print(f"Íµ¨Ï°∞ ÏòàÏ∏°Îêú ÏãúÌÄÄÏä§ Ïàò: {len(dots)} / Ï†ÑÏ≤¥: {len(df)}")

# Î∂ôÏù¥Í≥† Ï†ÄÏû•
df["structure"] = dots
df.to_csv("../data/processed/positive_structures.csv", index=False)
print("‚úÖ Íµ¨Ï°∞Í∞Ä Ìè¨Ìï®Îêú positive_structures.csv Ï†ÄÏû• ÏôÑÎ£å")
````
